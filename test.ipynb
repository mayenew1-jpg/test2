# 3.1 How many seconds are in an hour?

```python
seconds_in_hour = 60 * 60
print("3.1 Seconds in an hour:", seconds_in_hour)
```

# 3.2 Assign the result to a variable called seconds_per_hour

```python
seconds_per_hour = seconds_in_hour
print("3.2 seconds_per_hour =", seconds_per_hour)
```

# 3.3 How many seconds are in a day?

```python
seconds_in_day = seconds_per_hour * 24
print("3.3 Seconds in a day:", seconds_in_day)
```

# 3.4 Save the result in a variable called seconds_per_day

```python
seconds_per_day = seconds_in_day
print("3.4 seconds_per_day =", seconds_per_day)
```

# 3.5 Floating-point division

```python
floating_result = seconds_per_day / seconds_per_hour
print("3.5 Floating-point division:", floating_result)
```

# 3.6 Integer division

```python
integer_result = seconds_per_day // seconds_per_hour
print("3.6 Integer division:", integer_result)
```
